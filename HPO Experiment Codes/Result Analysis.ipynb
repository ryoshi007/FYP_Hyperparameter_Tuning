## 1.0 Analysis by Execution Time
This section will identify which method took the shortest time to complete the hyperparameter tuning processes.

```python
# Filtering out the 'Baseline' method
data_filtered = data[data['Hyperparameter_Tuning'] != 'Baseline']

# Grouping by 'ML_Model' and 'Hyperparameter_Tuning'
grouped_data = data_filtered.groupby(['ML_Model', 'Hyperparameter_Tuning'])

# Calculating average, min, and max time taken for each group
stats = grouped_data['Time_Taken'].agg(['mean', 'min', 'max'])

# Convert time to hours and minutes if it exceeds 1 hour (3600 seconds)
stats['mean_hrs_mins'] = stats['mean'].apply(lambda x: f"{int(x // 3600)}h {int((x % 3600) // 60)}m" if x >= 3600 else f"{x:.2f} s")
stats['min_hrs_mins'] = stats['min'].apply(lambda x: f"{int(x // 3600)}h {int((x % 3600) // 60)}m" if x >= 3600 else f"{x:.2f} s")
stats['max_hrs_mins'] = stats['max'].apply(lambda x: f"{int(x // 3600)}h {int((x % 3600) // 60)}m" if x >= 3600 else f"{x:.2f} s")

# Display the results
stats.reset_index()
```

## 2.0 Analysis By Best AUC_ROC Score
This section will search for the method that has the best auc-roc score

```python
# Filter to include the baseline in the analysis
data_filtered = data.copy()

# Define a function to get the top 2 methods based on ROC_AUC improvement
def get_top_2_methods(group):
    return group.sort_values('Testing_Improvement_ROC_AUC_(%)', ascending=False).head(2)

# Apply the function to each combination of seed and ML model
top_methods = data_filtered.groupby(['Seed_Value', 'ML_Model']).apply(get_top_2_methods)

# Reset index to make 'Seed_Value' and 'ML_Model' regular columns again
top_methods.reset_index(drop=True, inplace=True)

# Select the relevant details for the top 2 methods
top_methods_details = top_methods[[
    'Seed_Value', 'ML_Model', 'Hyperparameter_Tuning', 'Time_Taken',
    'Training_Accuracy', 'Improvement_Training_Accuracy(%)', 'Testing_Accuracy',
    'Testing_Improvement_Accuracy(%)', 'Testing_ROC_AUC', 'Testing_Improvement_ROC_AUC_(%)',
    'Best_Parameter_Found'
]]
```

## 4.0 Analysis by Diversity in Solutions
This section will find which method has the greatest diversity in its solution found. This is done by converting the categorical parameter into numerical values via one-hot encoding and performing min-max scaling on numerical parameters. After that, the average Manhattan distance differences between each hyperparameter solution represented by a point is calculated and plotted. Besides, these points will be reduced to 2-dimensions via PCA reduction and then plot on the graph.  

```python
experiment_data_new = pd.read_csv('Same Split Comparisons.csv')
```

```python
experiment_data_new.head()
```

```python
import ast

# Function to safely parse the 'Parameters' string into a dictionary
def safe_parse_parameters(param_str):
    try:
        # Replacing '|' with ',' to correct the dictionary format
        param_str = param_str.replace('|', ',')
        return ast.literal_eval(param_str)
    except:
        return None

# Apply the parsing function to the 'Parameters' column
experiment_data_new['Parameters'] = experiment_data_new['Parameters'].apply(safe_parse_parameters)

# Remove rows where parameters could not be parsed
experiment_data_new = experiment_data_new[experiment_data_new['Parameters'].notnull()]

# Displaying a few examples of parsed parameters
experiment_data_new[['Model', 'Method', 'Parameters']].head()
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import numpy as np
from itertools import product
```

### 4.1 Logistic Regression

```python
# Selecting data for Logistic Regression
lr_data = experiment_data_new[experiment_data_new['Model'] == 'Logistic Regression']

lr_data.tail()
```

```python
# Extracting parameters
lr_params = list(lr_data['Parameters'])

numerical_params_lr = ['C', 'l1_ratio']
categorical_params_lr = ['penalty', 'solver']

# Create lists to hold the separated values
numerical_values = {param: [] for param in numerical_params_lr}
categorical_values = {param: [] for param in categorical_params_lr}
```

```python
# Separate the parameter values
for params in lr_params:
    for param in numerical_params_lr:
        numerical_values[param].append(params.get(param, np.nan))  # Use NaN for missing values
    for param in categorical_params_lr:
        categorical_values[param].append(params.get(param, 'None'))  # Use 'None' for missing values
```

```python
lr_params
```

```python
# Normalize numerical parameters
scaler = MinMaxScaler()
for param in numerical_params_lr:
    values = np.array(numerical_values[param]).reshape(-1, 1)
    numerical_values[param] = scaler.fit_transform(values).flatten()
```

```python
# One-hot encode categorical parameters
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
for param in categorical_params_lr:
    values = np.array(categorical_values[param]).reshape(-1, 1)
    encoded = encoder.fit_transform(values)
    # Updating the dictionary with encoded values (each category becomes a separate key)
    categories = encoder.get_feature_names_out([param])
    for i, category in enumerate(categories):
        categorical_values[category] = encoded[:, i]
```

```python
# Combine numerical and categorical values into a final parameter representation
final_params = []
for i in range(len(lr_params)):
    combined = {param: numerical_values[param][i] for param in numerical_params_lr}
    combined.update({param: categorical_values[param][i] for param in categorical_values.keys()})
    final_params.append(combined)
```

```python
# Adjust final_params to include only numerical values (normalized and one-hot encoded)
final_params_adjusted = []
for param_set in final_params:
    # Replace NaN with 0 and include only numerical values
    param_set_numerical = {k: (0 if np.isnan(v) else v) for k, v in param_set.items() if isinstance(v, float)}
    final_params_adjusted.append(param_set_numerical)
```

```python
# Function to calculate the Manhattan distance excluding NaN values
def manhattan_distance(params1, params2):
    common_keys = set(params1.keys()).intersection(set(params2.keys()))
    distance = 0
    for key in common_keys:
        distance += abs(params1[key] - params2[key])
    return distance
```

```python
from collections import defaultdict

# Dictionary to store distances for each method
method_distances = defaultdict(list)

# Recalculate the Manhattan distances excluding NaN values
for method in lr_data['Method'].unique():
    method_indices = lr_data[lr_data['Method'] == method].index
    for i in method_indices:
        for j in method_indices:
            if i < j:  # Avoid duplicate pairs and self-comparison
                dist = manhattan_distance(final_params_adjusted[i], final_params_adjusted[j])
                method_distances[method].append(dist)
```

```python
import seaborn as sns

distances_list = [(method, distance) for method, distances in method_distances.items() for distance in distances]

# Create a DataFrame from the list of tuples
distances_df = pd.DataFrame(distances_list, columns=['Method', 'Manhattan Distance'])

# Calculate the average distance for each method
average_distances = distances_df.groupby('Method')['Manhattan Distance'].mean().reset_index()
average_distances.columns = ['Method', 'Average Manhattan Distance']

# Create the strip plot with jitter, specifying the order
plt.figure(figsize=(10, 6))
sns.stripplot(x='Method', y='Manhattan Distance', data=distances_df, jitter=True, order=method_order)

# Plot the average distances using a different color and marker style
# The DataFrame is already ordered by the 'Method' column in the same order as 'method_order'
sns.scatterplot(x='Method', y='Average Manhattan Distance', data=average_distances, 
                color='red', label='Average Distance', marker='o')

# Add titles and labels
plt.title('Dot Plot of Manhattan Distances with Average for Each Hyperparameter Tuning Method')
plt.xlabel('Hyperparameter Tuning Method')
plt.ylabel('Manhattan Distance')
plt.legend()

# Show the plot
plt.show()
```

```python
# save the information of dot plot
distances_df_for_csv = pd.DataFrame(distances_list, columns=['Method', 'Manhattan Distance'])

distances_df_for_csv.to_csv('lr_manhattan_distances.csv', index=False)
```

```python
pca_2d = PCA(n_components=2)
pca_result_2d = pca_2d.fit_transform([list(params.values()) for params in final_params_adjusted])
pca_1_2d = pca_result_2d[:, 0]
pca_2_2d = pca_result_2d[:, 1]

# Create a dataframe for PCA results and corresponding methods
pca_df = pd.DataFrame({
    'PCA_1': pca_1_2d,
    'PCA_2': pca_2_2d,
    'Method': lr_data['Method'].values  # Assuming the order hasn't changed
})

pca_df.to_csv('lr_pca_2d_results.csv', index=False)
```
